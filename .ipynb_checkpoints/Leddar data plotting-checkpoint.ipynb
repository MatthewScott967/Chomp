```python
%matplotlib inline
from __future__ import division
import serial
from collections import defaultdict
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
import xlsxwriter
import math
import collections
```

```python
ser = serial.Serial('/dev/tty.usbmodem1411', 115200)
```

```python
all_distances_by_seg = defaultdict(list)
for _ in range(10000):
    distances_by_seg = defaultdict(list)
    detections = ser.readline().rstrip().split("\t")
    for detection in detections:
        segment, distance = detection.split("/")
        distances_by_seg[int(segment)].append(int(distance))
    for segment, distances in distances_by_seg.iteritems():
        distances_by_seg[segment] += [np.nan] * (2 - len(distances))
    for segment, distances in distances_by_seg.iteritems():
        all_distances_by_seg[segment] += distances
    
```

```python
all_distances_by_seg = pd.DataFrame(all_distances_by_seg)
```

```python
fig, axes_array = plt.subplots(1, 16, figsize=(25, 20), subplot_kw={"axisbg": "white"})
fig.suptitle("Leddar returns", fontweight="bold", fontsize=24)
axes_array = axes_array.flatten()
for plot_num, segment in enumerate(all_distances_by_seg):
    ax = axes_array[plot_num]
    data = pd.Series(all_distances_by_seg[segment])
    ax.plot(data, range(len(data)), marker='.', markersize=2, linewidth=0)
    mean = data.mean()
    stdev = data.std()
    low_lim = int(math.floor(data.min()))
    high_lim = int(math.ceil(data.max()))
    ax.set_xlim([low_lim - stdev, high_lim + stdev])
    ax.xaxis.set_ticks([low_lim, high_lim])

    ax.set_title("{}".format(segment), fontsize=18, fontweight="bold")

for ax in axes_array:
    plt.setp(ax.get_xticklabels(), rotation=90, fontsize=16)

for ax in axes_array[1:]:
    plt.setp(ax.get_yticklabels(), visible=False)

axes_array[7].set_xlabel("             Distance", fontsize=20, fontweight="bold", labelpad=50)
axes_array[0].set_ylabel("Time", fontsize=20, fontweight="bold", labelpad=50)
plt.setp(axes_array[0].get_yticklabels(), fontsize=16)
    
plt.subplots_adjust(hspace=0.3, wspace=0.3)
```

```python
for plot_num, segment in enumerate(all_distances_by_seg):
    data = pd.Series(all_distances_by_seg[segment]).dropna()
    mean = data.mean()
    stdev = data.std()
    print segment, stdev / mean * 100
```

```python
workbook = xlsxwriter.Workbook("2016-02-24 Leddar data sample.xlsx")
cell_format = workbook.add_format()
cell_format.set_font_name('Consolas')
header = range(16)
worksheet = workbook.add_worksheet("data")
for col_num, field in enumerate(header):
    worksheet.write(0, col_num, field, cell_format)
row_num = 1
for col_num, (segment, detections) in enumerate(sorted(all_distances_by_seg.iteritems())):
    for row_num, detection in enumerate(detections, start=1):
        worksheet.write(row_num, col_num, str(detection), cell_format)
workbook.close()
```

```python
detections_by_time = defaultdict(str)
for segment, detections in sorted(all_distances_by_seg.iteritems()):
    for time, detection in enumerate(detections, start=1):
        detections_by_time[time] += "{}/{}\t".format(segment, detection)
with open("leddar_data.txt", 'w') as fh:
    for time, line in sorted(detections_by_time.iteritems()):
        fh.write(line + "\n")
```
