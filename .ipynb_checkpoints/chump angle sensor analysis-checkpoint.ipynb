```python
def polyfit_derivative(data, num_points):
    poly_derivatives = list()
    for i in range(len(data) - num_points):
        x = sum(range(num_points)) / num_points
        polyfit = np.polyfit(range(num_points), data[i:i + num_points], 2)
        first_derivative = 2 * polyfit[0] * x + polyfit[1]
        poly_derivatives.append(first_derivative * 1000)
#         print first_derivative
    return poly_derivatives
```

```python
angles = list()
pressures = list()
with open("/Users/legion/Dropbox/Chomp/chump_angle_pressure4.txt") as fh:
    for line in fh:
        line = line.rstrip().split("\t")
        if line[0] == "throw_close_time":
            throw_close_time = int(line[1])
        elif line[0] == "vent_open_time":
            vent_open_time = int(line[1])
        else:
            angle, pressure = map(int, line)
            angles.append(angle)
            pressures.append(pressure)
angles = pd.Series(angles)
pressures = pd.Series(pressures)
fig = plt.figure()
ax1 = plt.gca()
ax1.plot(angles)
plt.ylabel("angle from start (degrees)", color="blue")
plt.xlabel("time (ms)")
# ax1 = plt.gca()
ax2 = ax1.twinx()
ax2.plot(pressures, color="red")
plt.ylabel("pressure (psi)", color="red")
ax1.axvline(throw_close_time, linestyle="dashed", color="black")
ax1.axvline(vent_open_time, linestyle="dashed", color="black")
ymin1, ymax1 = ax1.get_ylim()
ax1.text(0, ymin1 - 15,'throw valve open', rotation=90, ha="center")
ax1.text(throw_close_time, ymin1 - 15,'throw valve close', rotation=90, ha="center")
ax1.text(vent_open_time, ymin1 - 15,'vent valve open', rotation=90, ha="center")
ymin2, ymax2 = ax2.get_ylim()
# ymin = min(ymin1, ymin2)
# ymax = max(ymax1, ymax2)
# ax1.set_ylim(-20, ymax)
ax2.set_ylim(ymin2, ymax2+10)
ax1.xaxis.grid(True, which="both")
minorLocator = AutoMinorLocator(10)
ax1.xaxis.set_minor_locator(minorLocator)
fig.patch.set_facecolor('white')
```

```python
# read data from file
angles = list()
pressures = list()
with open("/Users/legion/Dropbox/Chomp/chomp_angle_manualtest1.txt") as fh:
    for line in fh:
        line = line.rstrip().split("\t")
        if line[0] == "throw_close_time":
            throw_close_time = int(line[1])
        elif line[0] == "vent_open_time":
            vent_open_time = int(line[1])
        elif line[0] not in ("swing", "swung"):
            angle, pressure = map(int, line)
            angles.append(angle)
            pressures.append(pressure)

# set up x and y for plotting
angles = pd.Series(angles)
pressures = pd.Series(pressures)
timestep = 5  # in ms
x = np.arange(0, len(angles) * timestep, timestep)

# make plots
fig = plt.figure()
ax1 = plt.gca()
ax1.plot(x, angles)
plt.ylabel("angle from start (degrees)", color="blue")
plt.xlabel("time (ms)")
# ax1 = plt.gca()
ax2 = ax1.twinx()
ax2.plot(x, velocities, color="red")
plt.ylabel("pressure (psi)", color="red")

# valve action labels
ax1.axvline(throw_close_time * timestep, linestyle="dashed", color="black")
ax1.axvline(vent_open_time * timestep, linestyle="dashed", color="black")
ax1.text(0, ymin1 - 15,'throw valve open', rotation=90, ha="center")
ax1.text(throw_close_time * timestep, ymin1 - 15,'throw valve close', rotation=90, ha="center")
ax1.text(vent_open_time * timestep, ymin1 - 15,'vent valve open', rotation=90, ha="center")

# adjust axis limits
ymin1, ymax1 = ax1.get_ylim()
ymin2, ymax2 = ax2.get_ylim()
# ymin = min(ymin1, ymin2)
# ymax = max(ymax1, ymax2)
ax1.set_ylim(ymin1, ymax1)
ax2.set_ylim(ymin2, ymax2)

# adjust major tick location
ax1.yaxis.set_ticks(np.arange(0, angles.max() + 15, 15))

# adjust gridlines
ax1.yaxis.grid(b=True, which='major')
ax1.xaxis.grid(True, which="both")
minorLocator = AutoMinorLocator(10)
ax1.xaxis.set_minor_locator(minorLocator)
fig.patch.set_facecolor('white')  # make plot surround opaque
```
